```python
# Get a random number between 0 to 6 for Dice using random.randrange():

# 'snake_or_ladder' tells us if we got a ladder or snake on our new position :
# N - None;   S - Snake ;   L - Ladder 
snake_or_ladder = 'N'

# "x" holds the current position of our set piece
x = 0

# Game continues until the user reaches the Final Destination i.e x = 100

while(x < 100):
  print('\n\n Choose one of the following options : C -> Continue, Restart -> R, End -> E')
  a = input('\n\n Enter your option :')
  if(a == 'r'):
    x = 0
    print('\n\n Game Restarts ')
  elif (a == 'e'):
    print('\n\n Game Ends ')
    break
  else :
    print('\n\n Game Continues ')
  dice = random.randrange(1,6)
  print('\n\n Dice : ',dice)
  x = x + dice
  print('\n\n Went from {0} to {1}'.format(x-dice,x))
  next = get_next_position(x)
  if(next > x):
    snake_or_ladder = 'N'
    print('\n\n Damn! You just climbed a ladder and got up to position ',next)
  elif (next < x):
    snake_or_ladder = 'S'
    print('\n\n Ouchh! You just got bit by a snake ....you fell to position ',next)
  else :
    snake_or_ladder = 'N'
  x = next
  if(x == 100):
    print('Congrats ! You won the game')
    x = 0
  else:
    print('\n\n Your new position is ',x)
```
